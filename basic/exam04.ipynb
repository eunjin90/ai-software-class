숫자 3개를 입력받아 작은수에서 큰수순으로 출력

```python
a, b, c = map(int,input('숫자 3개를 입력하시오').split())

min = a
max = b

if a > b:
    min = b
    max = a

if c < min:
    print(f'{c} {min} {max}')
elif c < max:
    print(f'{min} {c} {max}')
else :
    print(f'{min} {max} {c}')

# 로딩중 숫자 3개를 입력하고 엔터
```
