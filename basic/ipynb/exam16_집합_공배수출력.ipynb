1 ~ 100사이의 3,5의 공배수 출력 

```python
muls = []
for i in range(1, 101):
    if i % 3 == 0 and i % 5 == 0:
        muls.append(i)
print(muls)
```

```python
muls = []
m = 3 * 5
for i in range(1, 101):
    if i % m == 0:
        muls.append(i)
print(muls)
```

```python
nums3 = {i for i in range(3, 101, 3)}
nums5 = {i for i in range(5, 101, 5)}
multipls = set(nums3 & nums5)
print(multipls)
print(min(multipls))
```
