```python
count = 0

for num in range(2, 101):   # 2부터 시작해서 하나씩 검사 # num: 지금 검사하는 숫자

    i = 2       # 2부터 나눠볼게!
    while i < num and num % i != 0:     # 나눠질 때까지 계속 i를 늘린다
        i += 1

    if i == num:
        print("%4d" % num, end=" ")
        count += 1

        if count % 5 == 0:
            print()
```

```python
count = 0

for num in range(2, 101):
    i = 2
    while i < num and num % i != 0:
        i += 1

    if i == num:
        print("%4d" % num, end='')
        count += 1

        if count % 5 == 0:
            print()
```

```python
start = int(input("시작단 : "))
end = int(input("종료단 : "))

for dan in range(start, end + 1):
    print("*** " + str(dan) + "단 ***")
    
    for num in range(1, 10):
        result = dan * num
        print(str(dan) + " * " + str(num) + " = " + str(result))
```

```python
dans = [2,3,4,5,6,7,8,9]

for i in range(0, len(dans), 3):

    # 단 제목
    for j in range(i, i+3):
        if j < len(dans):
            print("*** " + str(dans[j]) + "단 ***", end="\t")
    print()

    # 곱하기 부분
    for num in range(1,10):
        for j in range(i, i+3):
            if j < len(dans):
                print(str(dans[j]) + " * " + str(num) + " = %2d" % (dans[j]*num), end="\t")
        print()

    print()
```

```python
num = int(input("10이상 100미만의 숫자 입력 : "))

print("[1 ~", num, "까지의 369게임 시작]")

for i in range(1, num+1):

    clap = ""

    for j in str(i):
        if j == '3' or j == '6' or j == '9':
            clap = clap + "짝"

    if clap == "":
        print(i)
    else:
        print(clap)
```
