```python
errorCount = 0 # 에러개수
scoreTotal = 0 # 올바른 성적의 총합
for i in range(5) :
    score = int(input(f'{i+1}\'s 학생성적'))
    
    if 0 <= score <=100:
        if score >=90 : grade = 'A'
        elif score >= 80 : grade = 'B'
        elif score >= 70 : grade = 'C'
        elif score >= 60 : grade = 'D'
        else : grade = 'F'
    
        print(f'{score}성적의 학점은 {grade}입니다')
        scoreTotal = scoreTotal + score
    else :
        print('ErroR!!!')
        errorCount += 1
print(f'5회 입력 중 에러{errorCount}회 발생!!!')
print(f'성적의 총합 : {scoreTotal}')
print(f'성적의 평균 : {scoreTotal / 5}')
```

```python
# 소수판단 : 1과 자신의 정수로만 약수인 양의 정수 (약수의 개수가 2개)
for j in range(3):
    num = int(input('2 ~ 100사이의 정수 입력'))
    count = 0 # 약수의 개수를 세는 변수
    for i in range(1, num+1):
        if num % i == 0 : count +=1

    if count == 2: print(f'{num} : 소수입니다')
    else: print(f'{num} : 소수가 아닙니다')
```
