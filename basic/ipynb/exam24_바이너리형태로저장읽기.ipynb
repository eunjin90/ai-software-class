```python
import pickle   # pickle : 파이썬 데이터를 파일로 저장했다가 그대로 복원하는 기능
```

```python
a = [1,2,3,4,5]

with open("test01.pickle", "wb") as file:   # 바이너리 형태로 저장 (write binary)
    # pickling
    pickle.dump(a, file)    # file 라는곳에 a 객체를 저장하고싶다   (dump : 데이터를 파일에 넣는다)
print('파일저장 완료')
```

```python
with open("test01.pickle", "rb") as file:   # 읽기 (read binary)
    b = pickle.load(file)   # (load: 파일에서 데이터를 꺼낸다)
print('로드 완료')
print(b, type(b))
```

```python
def getTotal(*args):    # 숫자들 받아서 합계 구하는 함수
    s = 0
    for data in args:
        s += data
    return s

f = getTotal    # 함수도 변수처럼 다룰 수 있음
print(f(1,2,3,4))   # f는 이제 getTotal이랑 똑같음
```

```python
def prnTotal(*args):    # 합계 구해서 출력하는 함수
    s = getTotal(*args)
    print(f'{args} 총합 : {s}')
```

```python
with open('test02.pickle', 'wb') as file:
    pickle.dump(getTotal, file) # 함수 자체를 저장함
    pickle.dump(prnTotal, file)
print('파일 저장 완료...')
```

```python
import pickle
with open('test02.pickle', 'rb') as file:
    func01 = pickle.load(file)  # 파일에서 함수 꺼냄
    func02 = pickle.load(file)
print('파일 로드 완료...')
print(type(func01), type(func02))   # 진짜 함수로 복원됨
```

```python
func01(*[10,20,30,40,50])
func02(*[10,20,30,40,50])
```
