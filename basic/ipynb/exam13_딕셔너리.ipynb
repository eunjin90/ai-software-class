### 딕셔너리 내장 함수
- setdefault()  # 값을 넣을때 : key가 없으면 들어가고 key가 없으면 값이 들어가지않는다 그대로(중복허용 안할때 기존값 그대로)
- update()  # 값을 수정할때 : key 존재하면 value 수정, key 존재하지 않으면 새로운 데이터 삽입/ 덮어쓴다(중복허용은 안되고 덮어쓸때 쓴다)
- pop() : 벨류 하나만
- popitem() : 키벨류의 값
- clear() : 전부다 지움
- get(key) : key 에 대한 벨류값을 뽑아낼때 쓴다
- items() : 어떤 키 벨류의 쌍을 가지고 있는지 볼때
- keys() : 어떤 키들을 가지고있는지
- values() : 어떤 벨류들을 가지고있는지
- dict.fromkeys()

```python
x = {'a':10, 'b':20, 'c':30, 'd':40}
x
```

```python
x.setdefault('e',50)
print(x)
```

```python
x.setdefault('f')   # 벨류값을 안넣으면 None 표시
print(x)
```

```python
# x['b'] = 300
x.update(b=300)     # key는 문자열만 가능
print(x)
```

```python
x.update({1:"one"})     # 숫자는 이렇게
x.update([[2,"two"]])
x.update([[2,"TWO"]])
x.update([[2,"two"], [3, "three"]])
x.update([[5, "five"]])
print(x)
```

```python
# 300 값을 가진 item 삭제
x = {'a':100, 'b':200, 'c':300, 'd':400, 'e':300}
new_x = dict()
for key, value in x.items():
    if value != 300:
        new_x.setdefault(key, value)
x = new_x
print(x)
```
