시퀀스자료형: 여러 개의 데이터를 순서대로 저장하는 자료형 (연속적)
    가장많이쓰는 자료형 문자열(str)-"python" 연속적인공간을 만든다,
    리스트(list)-['A',2,3], 여러 데이터를 저장 수정 가능 (데이터를처리하는 자료형)
    튜플(tuple)-('A',2,3), 여러 데이터를 저장 수정 불가
    range-range(1,10), 숫자 범위

```python
# range

for i in range(5): print(i,end=' ')
print()
print(range(5))
```

```python
# list

a = list(range(5)) # 집합의 형태(빌라로 생각하면 쉬움),무한으로 값 넣을수있다
print(a)

for i in [1, 2, 3, 4, 5]:
    print(i,end=' ')
```

```python
a = [10, 20, 'A', 10.45, True, 'hello']
print(a, type(a))

# a라는 집합에 속해있는 원소인지 알아보는 방법
print(20 in a)
print(30 in a)

# a집합에 추가하기
a.append('add')
print(a)

# 0부터 시작(데이터베이스에서는 1부터시작, 프로그래밍에서는 0부터시작)
print(a[0],a[1])

# a집합에서 삭제하기
a.remove(20)
print(a)

a[3] = 12.56
print(a[3])
print(a)
```

```python
# tuple

# a = tuple(range(5))
a = (10, 20, 30, 'A')
print(a)

# a.append('add') 쓸 수 없다(수정 불가)
# a[0] = 1000

print(a, a[0])
```
