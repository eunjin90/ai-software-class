```python
# 비밀번호 변경서비스
members = {'aaa': '1111', 'bbb':'2222', 'ccc':'3333', 'ddd':'4444'}
members
```

***패스워드 변경서비스***
아이디를 입력하세요 : eee
아이디[eee]는 존재하지 않습니다
변경서비스를 종료합니다.

***패스워드 변경서비스***
아이디를 입력하세요:bbb
현재 패스워드를 입력하세요:2233
패스워드가 올바르지 않습니다.
변경서비스를 종료합니다.

***패스워드 변경서비스***
아이디를 입력하세요:bbb
현재 패스워드를 입력하세요 : 2222
변경할 패스워드를 입력하세요 : 5566
패스워드를 변경하였습니다.

<<<회원정보>>>
aaa 1111
bbb 5566
ccc 3333
ddd 4444

```python
print('*** 비밀번호 변경 서비스 ***')
id = input('아이디를 입력하세요')

if id not in members:
    print(f'아이디{id}는 존재하지 않습니다 변경서비스를 종료합니다')
    exit(0)

password = input('현재 패스워드를 입력하세요')
if members.get(id) != password:
    print('입력하신 패스워드가 올바르지 않습니다')
    exit(0)
```
