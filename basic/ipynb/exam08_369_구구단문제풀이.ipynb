3 6 9 게임

```python
#문자열방법
num = 41

for i in range(num):
    n = i+1
    # n = 28
    n1 = n % 10
    n10 = n // 10

    count = 0
    if n1 == 3 or n1 == 6 or n1 == 9 :
        count += 1
    if n10 == 3 or n10 == 6 or n10 == 9:
        count += 1
        print(f'{n} : {"짝" * count}')
```

```python
# 리스트방법
num = 41

for i in range(num):
    n = i+1
    # n = 28
    n1 = n % 10
    n10 = n // 10

    count = 0
    if n1 in [3, 6, 9] :
        count += 1
    if str(n10) in "369":
        count += 1
        print(f'{n} : {"짝" * count}')
    
    if n1 == 0:
        print(f'\t{"뽀"*n10}쑝')
```
