open(파일명, 파일모드)   파일객체를 연다     파일모드 "r" "w" (파일에 쓰기만 가능)
->
read()/write()          Read/write
->
close()                 파일객체를 닫는다

```python
import time
file = open("./data/test.txt","w")    # test.txt 파일을 ‘쓰기 모드(w)’로 열겠다 / 파일이 있으면 내용 전부 삭제(초기화)됨 / 파일이 없으면 새로 생성됨
file.write("Hello")
time.sleep(3)
file.close()
```

```python
file = open("./data/test.txt","r")
data = file.read()
file.close()
print(data)
```

```python
with open("./data/test.txt","w") as file:      # 자동으로 close() 됨 / as file은 변수 이름을 정해주는 것
    file.write("hello")

time.sleep(3)
print("완료")
```

```python
with open("./data/test.txt","r") as file:
    data = file.read()
print(data)
```
