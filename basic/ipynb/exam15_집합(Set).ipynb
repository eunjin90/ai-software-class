집합(Set) :
중복 (같은 값 하나만 저장됨)
순서 (인덱스 없음 → a[0] 이런 거 안됨)

진부분집합(<) : 완전히 작아야 함 (같으면 x), 부분집합(<=) : 같아도 포함됨
a = {1,2}
b = {1,2,3}
a < b   # True

a = {1,2}
b = {1,2}
a < b   # False (같으니까)  : 부분집합이면 true

a & b   # 교집합 → 둘 다 있는 값
a | b   # 합집합 → 전체 (중복 제거) , 공통값 포함
a - b   # 차집합 → a에는 있고 b에는 없는 값
a ^ b   # 대칭차집합 → 겹친 거 제거 , 공통값 제거

a.add(4)        # 하나 추가
a.update([5,6]) # 여러 개 추가
a.remove(2)     # 삭제

all() : 전부 True면 True , any() : 하나라도 True면 True

```python
# 교집합, 합집합, 차집합, 여집합 연산 수행
# 합집합
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
c = {5, 6, 7, 8}
# union : 집합들을 합쳐서 하나로 만드는 함수 (합집합)
print(f'합집합 : {set.union(a,b,c)}')
print(f'합집합 : {set.union(a,b)}')
print(f'합집합 : {(a|b)}')

# a랑 b를 합침 (중복은 자동 제거)
print(f'합집합 : {(a.union(b))}')
```

```python
# 교집합
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
c = {5, 6, 7, 8}
# intersection : 공통으로 들어있는 값만 뽑는 것
print(f'교집합 : {set.intersection(a,b)}')
print(f'교집합 : {a & b}')
print(f'교집합 : {a.intersection(b)}')
print(a)    # 원본은 안 바뀜
print(b)    # 원본은 안 바뀜
```

```python
# 차집합
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
# difference : 한쪽에만 있는 값
print(f'차집합 : {set.difference(a,b)}')
print(f'차집합 : {set.difference(b,a)}')
print(f'차집합 : {a - b}')
print(f'차집합 : {a.difference(b)}')
```

```python
# 여집합 : 전체에서 특정 집합을 뺀 것
# 여집합 = 전체 집합 - 특정 집합
# 대칭차집합 : 겹치는 거 빼고, 나머지
# 공통된 값은 제거
# 한쪽에만 있는 값만 남김
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
c = a.difference(b)
d = b.difference(a)
print(f'여집합 : {a-b |b-a}')
print(f'여집합 : {set.symmetric_difference(a,b)}')
print(f'여집합 : {a ^ b}')
print(f'여집합 : {a.symmetric_difference(b)}')
```

```python
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
# a = set.intersection_update(a,b)
# a = a.intersection(b)
# a = a & b         # 교집합 결과만 반환 (a 안 바뀜)
# set.intersection_update(a, b)
# a.intersection_update(b)
a &= b              # a를 교집합으로 변경 (a 바뀜), &= = 계산 + 자기 자신 변경
print(a)
print(b)
```

```python
a = {1, 2, 3, 4}
print(a)
# 원소 5를 삽입
# 합집합을 이용 : a |= {5}
a.update({5, 6, 3})  # 중복된거 제거하고 여러개의 데이터가 들어갈수있다
print(a)
```

```python
# 부분집합, 진부분집합
a = {1, 2, 3, 4}
print(f'부분집합 : {a <={1,2,3,4}}')
print(f'부분집합 : {a.issubset({1,2,3,4})}')
print(f'부분집합 : {a <={1,2,3,4,5}}')
print(f'부분집합 : {a.issubset({1,2,3,4,5})}')
print(f'부분집합 : {a < {1,2,3,4}}')
print(f'부분집합 : {a < {1,2,3,4,5}}')
```

```python

# 상위집합 (>=, issuperset(), 진상위집합(>))
a = {1, 2, 3, 4}
print(f'상위집합 : {a.issuperset({1,2,3})}')
print(f'상위집합 : {a.issuperset({1,2,3,4})}')
print(f'상위집합 : {a >= {1,2,3,4}}')
print(f'진상위집합 : {a > {1,2,3,4}}')
print(f'진상위집합 : {a > {1,2,3}}')
```

```python
# 세트사이의 교집합 원소 존재여부 판단
a = {1, 2, 3, 4}
print(a.isdisjoint({1,2,5,6}))
print(a.isdisjoint({5,6}))
```

```python
# == : 값 비교
# is : 주소(객체) 비교
a = {1, 2, 3, 4}
print(a == {1,2,3,4})
print(a == {4,3,2,1})
print(a == {4,3,2})
print(a != {4,3,2})
print(a is {1,2,3,4})
print(a is a)
print(a is b)
```

### set 조직함수
- add()
- remove() : 원소 삭제, 단 삭제할 원소가 없을 때 예외발생
- discard() : 원소 삭제, 삭제할 원소가 없어도 예외발생x
- pop()
- clear()
- copy()    : 깊은복사

```python
a = {1, 2, 3, 4}
a.add(5)
print(a)
len(a)
```

```python
a.remove(3)
a
# a.remove(3) 3을 삭제했는데 또 쓰면 에러발생
```

```python
print(a)
a.discard(2)
print(a)
a.discard(2) # 2를 삭제했는데 또 써도 에러발생하지않음
```

```python
a = {1, 2, 3, 4}
print(f'pop() : {a.pop()}')
print(a)
print(f'pop() : {a.pop()}')
print(a)
```
