```python
def print_book(title, writer, price):
    print('제목 : ', title)
    print('지은이 : ', writer)
    print('가격 : ', price)
```

```python
print_book('홍길동전','허균', 19000)
print_book(title='홍길동전',writer='허균',price=19000)
```

```python
book = {'title': '홍길동전', 'writer': '허균', 'price': 19000}
print(book)
print_book(**book)
print_book(*book)
```

```python
def prn_book(**kwargs):   #  키벨류쌍으로 kwargs로 쓰기도함(변수라 아무거나 써도되지만 통상)
    # print(kwargs)
    for key, value in kwargs.items():
        print(f'{key} : {value}')
```

```python
prn_book(title='홍길동전', writer='허균', price=19000)
prn_book(title='신데렐라', price=14000)
prn_book(**book)
```

문자열 함수
- replace(x, y) : x에 해당하는 문자열을 y로 대체
- split() : 문자열을 나눠서 리스트로 반환
- join(x) : 리스트를 문자열로 합침
- upper()
- lower()
- lstrip() : 왼쪽 공백 제거
- rstrip() : 오른쪽 공백 제거
- strip() : 양쪽 공백 제거
- center() : 가운데 정렬
- zfill() : 왼쪽 공백을 0으로 채움
- find(x) : x 문자열 위치 검색(없으면 -1 반환)
- index()
- count()

```python
def strRepalce(msg, before, after):
    c = ''
    for s in msg:
        if s == before:
            c += after
        else:
            c += s
    print(c)
```

```python
print(strRepalce('hello','l','rr'))
```
