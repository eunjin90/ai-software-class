```python
with open('test3.txt', 'w') as file:
    print('문자열을 입력하세요. quit 입력시 종료')
    while True:
        msg = input()
        if msg.lower() == "quit":
            break
        file.write(f'{msg}\n')
```

```python
with open('test3.txt', 'r') as file:
    data = file.read()
    print(data)
```

```python
# readline()
with open('test3.txt', 'r') as file:
    while True:
        line = file.readline().strip('\n')
        if line == "":
            break
        # line = line.strip('\n')
        print(line, len(line))
```

```python
with open('test3.txt', 'r') as file:
    for line in file:
        line = line.strip('\n')
        print(line)
```

```python
with open('test3.txt', 'r') as file:
    lines = file.readlines()
print(lines)

for line in lines:
    print(line.strip('\n'))
```
