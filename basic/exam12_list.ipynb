```python
a = list(range(1, 10))
```

```python
a
```

```python
a[0:5]
```

```python
print(a[0:len(a)])
```

```python
print(a[:4])   # a[0:4]
print(a[4:]) # a[4:len(a)]
print(a[4:-1]) # a[4:len(a)-1]
```

```python
print(a[2:7:2]) # 2번지부터 7번지까지 2개씩띄엄
print(a[2::2]) # 2번째부터 마지막까지 2개씩 # a[2:len(a):2]
print(a[::2]) # 맨 처음부터 끝까지 2씩증가
print(a[::-1])
```

```python
print(a[::-1])
print(a[8:1:-3])    # 8번지부터 1번지까지 3씩 감소
```
