```python
for count in range(1, 5+1, 3): # 0부터 시작되기때문에 1부터시작하고싶으면 이렇게
    #(스타트값(처음시작하는 값), 0부터 세기때문에 지정값+1을 해줌,스탭수)
    print('Hello',count)
```
