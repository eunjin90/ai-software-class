def 예약어를 활용한 함수 정의
    -함수 안으로 값(정보)을 전달 ->인자 라고 한다
    -함수가 보내는 값 ->변환 값 이라고한다

```python
a = ('*' *5)
print(a)

# 함수 정의 / 함수 선언
def printStar(n):
    print('*' * n, end='')
printStar(2)
```

```python
# 함수실행 / 함수호출
printStar(5)
print('Hello')
printStar(10)
print('Hi')
printStar(5)
print('Good-bye')
printStar(10)
```

```python
def printDash():
    print('-' * 5)
```

```python
def printChars(c):
    print(f'{c}' * 5)
```

```python
printChars('*')
print('Hello')
printChars('-')
print('Hi')
printChars('*')
print('Good-bye')
printChars('!')
```
