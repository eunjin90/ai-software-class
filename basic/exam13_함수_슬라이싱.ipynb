# list 함수
- append : 원소 추가
- insert : 특정위치에 원소 추가
- pop : 마지막 원소 또는 특정위치의 원소 삭제
- remove : 특정값을 가진 원소 삭제
- extend : 두개의 리스트를 한개의 리스트로 확정
- index : 특정값의 원소 위치를 알아낼때
- count : 리스트 내 특정값의 개수
- clear : 전체데이터 삭제

```python
a = [10, 20, 30]
print(a)
```

```python
a.append(40)
print(a, id(a))
```

```python
b = [100, 200]
# a.append(b)
a = a + b
print(a, id(a))
```

```python
a.extend(b)
print(a, id(a))
```

```python
a.insert(0, 55)
a
```

```python
a.insert(4, 'A')
a
```

```python
# a.append("hello")
a.insert(len(a), 'hello')
a.insert(16, 'bye')
a
```

```python
a = [10, 20, 30, 40, 50]
# a[2] = 1000
# a[2:2] = [1000, 2000] # 중간삽입할수있다
a[2:3] = [1000, 2000, 3000] # 30 데이터가 빠지면서 1000이 들어감
a
```

```python
#delNum = a.pop()
delNum = a.pop(3)
print(delNum)
print(a)
```

```python
a = [10, 20, 30, 40, 20, 40, 60]
a.remove(20)
a
```

```python
delNum = 1000
if a.count(delNum) :
    a.remove(delNum)
print(a)
```

```python
a = [10,20,30,40,50]
delIndex = 5
#if delIndex >= 0 and delIndex < len(a):
if 0 <= delIndex < len(a):
    a.pop(delIndex)
print(a)
```

```python
a = [10,20,30,40,50]
print(a)
a.reverse() # 거꾸로
print(a)
```

```python
a = [10,20,40,50,15,66]
print(a)
# b = sorted(a) # 정렬(원래 원본이 안바뀜) 원리시트의 내용이 변경되지 않음
b = sorted(a, reverse=True) # 디폴드값이 false, 오름차순정렬
print(a)
print(b)
```

```python
a = ["boat", "banana", "zoo", "apple"]
print(a)
print(sorted(a))
```

```python
a = [10, 20, 40, 3, 50, 15, 66, 9]
a = [str(data) for data in a]
print(a) # 리스트에 있는 모든 원소들을 문자로바꿈
print(sorted(a)) # 문자와 숫자는 정렬하는 방식이 달라진다
```

```python
a = [10, 20, 30, 40, 50]
print(a)
# 전체데이터 삭제
# while len(a):     # 와일문으로 pop이용해서 삭제
#     a.pop()
# a.clear()     # 클리어함수로 삭제
del a[:]    # 슬라이싱으로 삭제
print(a)
```

```python
a = [10, 20, 30, 40, 50]
# b = a     # shallow copy(얕은복사)
# b = a[:]    # deep copy(깊은복사)
b = a.copy()    # deep copy(깊은복사)

print(a, id(a))
print(b, id(b))

## == is
print(f'a == b : {a == b}')
print(f'a is b : {a is b}')

a[2] = 1000

print(a, id(a))
print(b, id(b)) # a의 배열을 복사한 b는 a의 값을 변경하더라도 b는 변경되지않는다
```
