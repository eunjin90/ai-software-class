```python
def getTotal(a = 0, b = 0, c = 0):  # 원래 인자가 3개가 필요한데 디폴트값을 주면 값을 다 적어주지않아도 계산됨 print(getTotal(10,20)) -> print(getTotal(10,20,0))
    sum = a + b + c
    return sum
```

```python
print(getTotal(1, 2, 3))
print(getTotal(10, 20))  # 에러 인자가 3개인데 디폴드값을 0으로 정해줘서 10 + 20 + 0
print(getTotal(5))
print(getTotal())

print(getTotal(10, 20, 30))
# total = getTotal(10, 20, 30)
# print(total)
```
